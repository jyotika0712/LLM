```python
from torchtext.datasets import AG_NEWS
from torchtext.data import to_map_style_dataset

train_set = AG_NEWS(split='train')
train_iter = iter(train_set)
train_set_map = to_map_style_dataset(train_set)

test_set=AG_NEWS(split='test')
test_iter = iter(test_set)
test_set_map=to_map_style_dataset(test_set)
```

```python
#DATASET SIZES
print(f"Training set size: {len(train_set_map)}")
print(f"Test set size: {len(test_set_map)}")
```

```python
#UNIQUE LABELS
labels_set={label for label,_ in train_set_map}
print(f" the set if Unique labels: {labels_set}")
```

```python
labels_list=[label for label,_ in train_set_map]
labels_tensor=torch.tensor(labels_list)
```

```python
unique_labels, counts = torch.unique(labels_tensor, return_counts=True)
label_dict={1:'World',2:'Sports',3:'Business',4:'Science & Technology'}
print_sample_size=2
for label in unique_labels:
  count=0
  print(f"some samples from label {label},corresponding to category\"{label_dict[int(label)]}\"")
  for l,t in train_set_map:
    if l==label:
      print(t)
      count+=1
      print(" ")
      if count>=print_sample_size:
        break
  print("-------------------------------------------------------------------")
```

```python
#Plotting the histogram
import numpy
plt.bar(unique_labels.tolist(), counts.tolist(), tick_label=list(range(1,5)))
plt.xlabel("label")
plt.ylabel("Count")
plt.title("Label Distribution in training data")
plt.show()
```
